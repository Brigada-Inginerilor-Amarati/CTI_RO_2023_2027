# Exercitiul 1

Create tensors of the following ranks: 1, 2, and 3. Use values between 0 and 1 to fill your tensors. The size of the tensors can be defined as you wish, given that the ranks are created correctly. Print the shape of each of the tensors using the shape property.

```python
import torch

# in acest caz -> generam un vector pe o linie cu 3 coloane
tensorRankOne = torch.rand(3)

print(tensorRankOne)
print(tensorRankOne.shape)

# primul parametru -> linii, al doilea -> coloane
tensorRankTwo = torch.rand(2, 3)

print(tensorRankTwo)
print(tensorRankTwo.shape)

tensorRankThree = torch.rand(2, 3, 4)

print(tensorRankThree)
print(tensorRankThree.shape)
```

# Exercitiul 2

The sigmoid activation function is an essential function in neural networks and is defined as follows:$$\text{sigmoid}(x) = \frac{1}{1 + e^{-x}}$$Take the derivative of this function with respect to $x$, where $x = [-10, 10, 8]$.

```python
def sigmoid(x):
  return 1 / (1 + torch.exp(-x))

x = torch.tensor([-10.0, 10.0, 8.0], requires_grad = True)

y = sigmoid(x)

y.backward(torch.ones_like(x))

print("Derivative of sigmoid at x = ", x.tolist(), " is ", x.grad.tolist())

# sau mai putem face

x = torch.tensor([-10.0, 10.0, 8.0], requires_grad = True)

y = sigmoid(x)

y.sum().backward()

print("Derivative of sigmoid at x = ", x.tolist(), " is ", x.grad.tolist())
```

# Exercitiul 3

Define the following two matrices:

$$A = \begin{bmatrix} 1 & 2 & -3 \\ 4 & 5 & 10 \end{bmatrix}, B = \begin{bmatrix} 10 & 11 & -2 \\ 13 & -3 & 8 \end{bmatrix}$$
calculate the Hadamard product of the two matrices;add the scalar $a = 10$ to each element of matrix $B$;print the transpose of both matrices.

```python
A = torch.tensor(
    [
        [1, 2, -3],
        [4, 5, 10]
    ]
)

B = torch.tensor(
    [
        [10, 11, -2],
        [13, -3, 8]
    ]
)

hadamardProduct = A * B

print("Hadamard Product is : ", hadamardProduct)

a = 10

B_new = B + a

print("B_new is : ", B_new)

print("Transpose of A is : ", A.T)
print("Transpose of B is : ", B.T)
```

# Exercitiul 4

Create the following two matrices:$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}, B = \begin{bmatrix} 10 & 11 \\ -2 & 0 \\ 1 & -3 \end{bmatrix}$$Multiply the two matrices and compute the Frobenius norm of $A$ and $B$.

```python
A = torch.tensor(
    [
        [1, 2, 3],
        [4, 5, 6]
    ],
    dtype = torch.float32
)

B = torch.tensor(
    [
        [10, 11],
        [-2, 0],
        [1, -3]
    ],
    dtype = torch.float32
)

product = torch.mm(A, B)

print("Product of matrix A and B is : ", product)

frobeniusNormA = torch.norm(A, p = "fro")
frobeniusNormB = torch.norm(B, p = "fro")

print("Frobenius norm of A : ", frobeniusNormA.item())
print("Frobenius norm of B : ", frobeniusNormB.item())
```
