## Exercise 2

The sigmoid activation function is an essential function in neural networks and is defined as follows:

$$sigmoid(x) = \frac{1}{1 + e^{-x}}$$

Take the derivative of this function with respect to x, where x = [-10, 10, 8].


```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))


x = torch.tensor([-10.0, 10.0, 8.0], requires_grad=True)

y = sigmoid(x)

y.backward(torch.ones_like(x))

print("Derivative of sigmoid at x = ", x, " is ", x.grad)
```

## Exercise 4

Create the following two matrices:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}$$

$$B = \begin{bmatrix} 10 & 11 \\ -2 & 0 \\ 1 & -3 \end{bmatrix}$$

Multiply the two matrices and compute the Frobenius norm of A and B.


```python
A = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
B = torch.tensor([[10, 11], [-2, 0], [1, -3]], dtype=torch.float32)

product = A @ B

fro_norm_A = torch.norm(A, p="fro")
fro_norm_B = torch.norm(B, p="fro")

print(product)
print(fro_norm_A.item())
print(fro_norm_B.item())
```
