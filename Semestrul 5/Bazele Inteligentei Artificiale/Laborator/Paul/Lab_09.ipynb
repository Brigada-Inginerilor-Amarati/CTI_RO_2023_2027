```python
import torch
from torch import nn
from torch.nn import functional as F
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
class ConvolutionalBlock(nn.Module):
    """The Residual block of ResNet."""

    def __init__(self, input_channels):
        super().__init__()
        self.conv3x3 = nn.Conv2d(input_channels, 32, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.conv1x1 = nn.Conv2d(32, 3, kernel_size=1)
        self.batchnorm = nn.BatchNorm2d(3)

    def forward(self, X):
        branch1 = X

        branch2 = self.conv3x3(X)
        branch2 = self.relu(branch2)
        branch2 = self.conv1x1(branch2)
        branch2 = self.batchnorm(branch2)
        branch2 = self.relu(branch2)

        return branch1 + branch2
```

```python
net = nn.Sequential(ConvolutionalBlock(3), nn.Flatten(), nn.Linear(3 * 32 * 32, 10))
net
```
